```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
epochs = [1, 2, 3, 4, 5]
single_machine_accuracy = [0.75, 0.80, 0.85, 0.88, 0.90]
distributed_accuracy = [0.75, 0.82, 0.87, 0.92, 0.95]

# Plotting the accuracy over epochs
plt.figure(figsize=(10, 6))
plt.plot(epochs, single_machine_accuracy, marker='o', label='Single Machine')
plt.plot(epochs, distributed_accuracy, marker='o', label='Distributed Learning')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Performance Comparison: Single vs Distributed Deep Learning')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
